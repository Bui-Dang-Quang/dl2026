```python
w0 = int(input("Enter the coefficient (a): "))
w1 = int(input("Enter the intercept (b): "))


def f(x):
    return w0 * x + w1

def f_derivative_w0(x):
    return x

# Gradient Descent function
def gradient_descent(starting_point, learning_rate, iterations):
    x = starting_point
    for i in range(iterations):
        grad = f_derivative_w0(x)
        x = x - learning_rate * grad
        print(f"Iteration {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")
    return x

# Run gradient descent
minimum = gradient_descent(starting_point=1, learning_rate=0.1, iterations=10)
print(f"\nLocal minimum at x = {minimum:.4f}, f(x) = {f(minimum):.4f}")


```

```python
# Lets try x power 2
def f(x):
    return x ** 2

def f_derivative_w0(x):
    return 2 * x

# Gradient Descent function
def gradient_descent_power_2(starting_point, learning_rate, iterations):
    x = starting_point
    for i in range(iterations):
        grad = f_derivative_w0(x)
        x = x - learning_rate * grad
        print(f"Iteration {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")
    return x

# Run gradient descent for x^2
minimum_power_2 = gradient_descent_power_2(starting_point=1, learning_rate=0.1, iterations=10)
print(f"\nLocal minimum for x^2 at x = {minimum_power_2:.4f}, f(x) = {f(minimum_power_2):.4f}")
```
